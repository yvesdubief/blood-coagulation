```python
%matplotlib inline
%config InlineBackend.figure_format='svg'

import numpy as np
import scipy.integrate as integrate
import matplotlib.pyplot as plt #calls the plotting library hereafter referred as to plt

def dy_dt(t, y):

    #kinetic constants k_n
    nspecies = len(y)
    nrates = 26
    k = np.zeros(nrates+1)
    k[0] = 3.2e6 #M-1 sec-1
    k[1] = 2.3e7 #M-1 sec-1
    k[2] = 4.4e5 #M-1 sec-1
    k[3] = 1.3e7 #M-1 sec-1
    k[4] = 2.3e4 #M-1 sec-1
    k[5] = 2.5e7 #M-1 sec-1
    k[6] = 2.2e7 #M-1 sec-1
    k[7] = 1.0e7 #M-1 sec-1
    k[8] = 7.5e3 #M-1 sec-1
    k[9] = 2.0e7 #M-1 sec-1
    k[10] = 1.0e7 #M-1 sec-1
    k[11] = 1.0e8 #M-1 sec-1
    k[12] = 6.0e-3 #sec-1
    k[13] = 1.0e-3 #sec-1
    k[14] = 1.0e-3 #sec-1
    k[15] = 2.0e7 #M-1 sec-1
    k[16] = 4.0e8 #M-1 sec-1
    k[17] = 1.0e8 #M-1 sec-1
    k[18] = 1.5e7 #M-1 sec-1
    k[19] = 9.0e5 #M-1 sec-1
    k[20] = 3.2e8 #M-1 sec-1
    k[21] = 5.0e7 #M-1 sec-1
    k[22] = 1.5e3 #M-1 sec-1
    k[23] = 7.1e3 #M-1 sec-1
    k[24] = 4.9e2 #M-1 sec-1
    k[25] = 7.1e3 #M-1 sec-1
    k[26] = 2.3e2 #M-1 sec-1
    
    #species
    TF = y[0]
    VII = y[1]
    TF_VII = y[2]
    VIIa = y[3]
    TF_VIIa = y[4]
    Xa = y[5]
    IIa = y[6]
    X = y[7]
    TF_VIIa_X = y[8]
    TF_VIIa_Xa = y[9]
    IX = y[10]
    TF_VIIa_IX = y[11]
    IXa = y[12]
    II = y[13]
    VIII = y[14]
    VIIIa = y[15]
    IXa_VIIIa = y[16]
    IXa_VIIIa_X = y[17]
    VIIIa1_L = y[18]
    VIIIa2 = y[19]
    V = y[20]
    Va = y[21]
    Xa_Va = y[22]
    Xa_Va_II = y[23]
    mIIa = y[24]
    TFPI = y[25]
    Xa_TFPI = y[26]
    TF_VIIa_Xa_TFPI = y[27]
    ATIII = y[28]
    mIIa_ATIII = y[29]
    IXa_ATIII = y[30]
    IIa_ATIII = y[31]
    TF_VIIa_ATIII = y[32]
    
    #reactions
    dydt = np.zeros(nspecies,1)
    
    #TF
    dydt[0] = -k[0]*TF*VII + k[0]*TF_VII \
                - k[1]*TF*VIIa + k[1]*TF_VIIa 
    #VII
    dydt[1] = -k[0]*TF*VII + k[0]*TF_VII \
                -k[2]*TF_VIIa*VII + k[2]*TF_VIIa*VIIa\
                -k[3]*Xa*VII + k[3]*Xa*VIIa\
                -k[4]*IIa*VII + k[4]*IIa*VIIa
    #VIIa
    dydt[2] = -k[1]*TF*VIIa + k[1]*TF_VIIa\
                -k[2]*TF_VIIa*VII + k[2]*TF_VIIa*VIIa\
                -k[3]*Xa*VII + k[3]*Xa*VIIa\
                -k[4]*IIa*VII + k[4]*IIa*VIIa
    #TF_VII
    dydt[3] = -k[0]*TF_VII + k[0]*TF*VII
    
    #TF_VIIa
    dydt[4] = -k[1]*TF_VIIa + k[1]*TF*VIIa\
                -k[2]*TF_VIIa*VII + k[2]*TF_VIIa*VIIa\
                -k[5]*TF_VIIa*X + k[5]*TF_VIIa_X\
                -k[6]*TF_VIIa*Xa + k[6]*TF_VIIa_Xa\
                -k[7]*TF_VIIa*IX + k[7]*TF_VIIa_IX\
                -k[21]*TF_VIIa*Xa_TFPI - k[21]*TF_VIIa_Xa_TFPI\
                -k[26]*TF_VIIa*ATIII + k[26]*TF_VIIa_ATIII
    #Xa
    dydt[5] = -k[3]*VII*Xa + k[3]*Xa*VIIa\
                -k[8]*II*Xa + k[8]*IIa*Xa\
                -k[16]*Xa*Va + k[16]*Xa_Va\
                -k[19]*Xa*TFPI + k[19]*Xa_TFPI\
                -k[23]*Xa*ATIII + k[23]*Xa_ATIII
    #X
    dydt[6] = -k[5]*TF_VIIa*X + k[5]*TF_VIIa_X\
                -k[5]*TF_VIIa_X + k[6]*TF_VIIa_Xa\
                -k[11]*IXa_VIIIa*X + k[11]*IXa_VIIIa_X\
                -k[11]*IXa_VIIIa_X + k[11]*IXa_VIIIa*Xa\
                -k[13]*IXa_VIIIa_X + k[13]*VIIIa1_L*VIIIa2*X*IXa
    #IIa
    
    return dydt

backend = 'vode'
r = integrate.ode(dy_dt).set_integrator(backend,max_step=1e-2, rtol=1.0e-6 , method='bdf', order=3, nsteps=10000)
# Part 1:
ICs = np.array([\
                #TF = y[0]
                5e-12, \#M
                #VII = y[1]
                1.00e-8, \ #M
                #TF_VII = y[2]
                0.0,\
                #VIIa = y[3]
                1.00e-10,\
                #TF_VIIa = y[4]
                0.0,\
                #Xa = y[5]
                0.0,\
                #IIa = y[6]
                0.0,\
                #X = y[7]
                1.60e-7,\
                #TF_VIIa_X = y[8]
                0.0,\
                #TF_VIIa_Xa = y[9]
                0.0,\
                #IX = y[10]
                9.00e-8,\
                #TF_VIIa_IX = y[11]
                0.0,\
                #IXa = y[12]
                0.0,\
                #II = y[13]
                1.40e-6,\
                #VIII = y[14]
                7.00e-10,\
                #VIIIa = y[15]
                0.0,\
                #IXa_VIIIa = y[16]
                0.0,\
                #IXa_VIIIa_X = y[17]
                0.0,\
                #VIIIa1_L = y[18]
                0.0,\
                #VIIIa2 = y[19]
                0.0,\
                #V = y[20]
                2.00e-8,\
                #Va = y[21]
                0.0,\
                #Xa_Va = y[22]
                0.0,\
                #Xa_Va_II = y[23]
                0.0,\
                #mIIa = y[24]
                0.0,\
                #TFPI = y[25]
                2.50e-9,\
                #Xa_TFPI = y[26]
                0.0,\
                #TF_VIIa_Xa_TFPI = y[27]
                0.0,\
                #ATIII = y[28]
                3.40e-6,\
                #mIIa_ATIII = y[29]
                0.0,\
                #IXa_ATIII = y[30]
                0.0,\
                #IIa_ATIII = y[31]
                0.0,\
                #TF_VIIa_ATIII = y[32]
                0.0])
y = np.copy(ICs)
t_0 = 0
t_final = 200

# Create vectors to store the solutions in; 
#     add extra space for intial condition

time = []
X = []
Y = []
Z = []

r.set_initial_value(ICs, t_0)

i = 1
while r.successful() and r.t < t_final:
    r.integrate(t_final,step=True)
    time.append(r.t)
    
    X.append(r.y[0])
    Y.append(r.y[1])
    Z.append(r.y[2])
    i+=1
    
print(i-1)

               
```
